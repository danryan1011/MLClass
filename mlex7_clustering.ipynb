```python
def findClosestCentroids(X, centroids):    results = []    for x in X:        costs = []        for c in centroids:            d = sum((x-c)**2)            costs.append(d)        results.append(costs.index(min(costs)))    return np.array(results)                
```

```python
def computeCentroids(X, idx, K):    centroids = []    for k in range(K):        k_bool = idx == k        centroids.append(np.mean(X[k_bool], axis=0))    return centroids#computeCentroids(X, t, 3)
```

```python
def runkMeans(X, initial_centroids, max_iters):    new_centroids = initial_centroids    K = len(initial_centroids)    for i in range(max_iters):        idx = findClosestCentroids(X, new_centroids)        new_centroids = computeCentroids(X, idx, K)    return new_centroids, findClosestCentroids(X, new_centroids)runkMeans(X, initial_centroids, 10)        
```

```python
def kMeansInitCentroids(X, K):    X_copy = X.copy()    np.random.shuffle(X_copy)    return X_copy[:K]    
```
