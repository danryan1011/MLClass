```python
data3 = scipy.io.loadmat('ex6data3')X3 = data3['X']y3 = data3['y'].T[0]X3val = data3['Xval']y3val = data3['yval'].T[0]
```

```python
def trainAndTest(X, y, Xval, yval, C_vec, sigma_vec):        results = []    for i in C_vec:        for j in sigma_vec:            machine = svm.SVC()            machine.C = i            machine.gamma = 1/(2*j**2)            machine.fit(X,y)            #print machine.predict(Xval)            accuracy = 1-sum((machine.predict(Xval)-yval)**2)/len(yval)            results.append(accuracy)    max_index = np.argmax(np.array(results))    C_max = C_vec[max_index / len(sigma_vec)]    sigma_max = sigma_vec[max_index % len(sigma_vec)]        return C_max, sigma_max, results        
```

```python
# Find C and sigma that result in maximum accuracyC3, sigma3, _ = trainAndTest(X3, y3, X3val, y3val, C_vec, C_vec)svm3 =svm.SVC()svm3.C = C3svm3.gamma = 1/(2*sigma3**2)svm3.fit(X3,y3)
```
