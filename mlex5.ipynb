```python
def linearRegCostFunction(X, y, theta, lmd):    m = len(y)*1.    J = (1./(2*m))*sum((theta.dot(X.T) - y)**2) + lmd/(2*m)*sum(theta[1:]**2)        grad = (1./m)*(theta.dot(X.T) - y).dot(X)    reg_term = lmd/m*theta    reg_term.T[0] *= 0    grad += reg_term        return J, grad    
```

```python
def polyFeatures(X, p):    #X should be a vector    X_poly = np.ones((len(X), p + 1))    for i in range(len(X)):        for j in range(p+1):            X_poly[i , j] *= X[i]**j    return X_poly
```

```python
def featureNormalize(X):    mu = np.array([np.mean(X[:,i]) for i in range(X.shape[1])])    sigma = np.array([np.std(X[:,i]) for i in range(X.shape[1])])    return (X - mu)/sigma, mu, sigma
```
