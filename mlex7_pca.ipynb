```python
def featureNormalize(X):    mu = np.mean(X, axis=0)    sd = np.std(X, axis=0)    return (X-mu)/sd
```

```python
X = featureNormalize(X)
```

```python
def pca(X):    m = len(X)    sigma = X.T.dot(X)    U, S, V = linalg.svd(sigma)    return U, S
```

```python
U, S = pca(X)
```

```python
def projectData(X, U, K):    U_reduce = U[:, :K]    Z = []    for i in range(len(X)):        Z.append(U_reduce.T.dot(X[i]))    return np.array(Z)    
```
